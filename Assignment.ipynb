**Iris data**

```python
# Impoert the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

Loading dataset

```python
data_set1 = pd.read_csv("Iris.csv")
data_set1.head()
```

Distribution of sepal lengths and widths among the Iris species

```python
# Distribution plot for Sepal Length
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(data=data_set1, x="SepalLengthCm", hue="Species", element="step", stat="density", common_norm=False)
plt.title('Distribution of Sepal Lengths')

# Distribution plot for Sepal Width
plt.subplot(1, 2, 2)
sns.histplot(data=data_set1, x="SepalWidthCm", hue="Species", element="step", stat="density", common_norm=False)
plt.title('Distribution of Sepal Widths')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import altair as alt
chart = alt.Chart(data_set1).mark_point().encode(
    alt.X("SepalLengthCm", title="Sepal Length (cm)"),
    alt.Y("SepalWidthCm", title="Sepal Width (cm)"),
    color="Species",
).properties(
    title="Distribution of Sepal Length and Width by Iris Species"
)
chart
```

Petal lenght accross different irsis species.

```python
# Plotting petal length distribution across species
plt.figure(figsize=(10, 6))
sns.boxplot(x='Species', y='PetalLengthCm', data=data_set1)
plt.title('Petal Length Variation Across Iris Species')
plt.xlabel('Species')
plt.ylabel('PetalLengthCm')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Species', y='PetalLengthCm', data=data_set1)
plt.title('Petal Length Distribution among Iris Species')
plt.xlabel('Species')
plt.ylabel('PetalLengthCm')
plt.show()
```

Scatter Plot of sepal lenght vs petal lenght

```python
sns.scatterplot(data=data_set1, x='SepalLengthCm', y='PetalLengthCm', hue='Species')
plt.title('Relationship between Sepal Length and Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
correlation = data_set1[['SepalLengthCm', 'PetalLengthCm']].corr().iloc[0, 1]
print(f'Correlation coefficient between Sepal Length and Petal Length: {correlation}')
```

Pairwise relationships between features
(sepal length, sepal width, petal length, petal width)

```python
sns.pairplot(data_set1, hue='Species', markers=["o", "s", "D"])
plt.show()
```

```python
#Creating dummy variable for Species column
data_cleaned1=pd.get_dummies(data_set1,columns=['Species'],dtype=int)
```

```python
correlation_matrix = data_cleaned1.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

Diabetes data

```python
data_set2 = pd.read_csv("diabetes.csv")
data_set2.head()
```

```python
diabetic_data = data_set2[data_set2['Outcome'] == 1]
```

```python
plt.figure(figsize=(5, 3))
sns.histplot(diabetic_data['Age'], kde=True)
plt.title('Age Distribution of Diabetic Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.histplot(diabetic_data['BMI'], kde=True)
plt.title('BMI Distribution of Diabetic Patients')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 3))

# Age Distribution
plt.subplot(1, 2, 1)
sns.histplot(diabetic_data['Age'], kde=True)
plt.title('Age Distribution of Diabetic Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')

# BMI Distribution
plt.subplot(1, 2, 2)
sns.histplot(diabetic_data['BMI'], kde=True)
plt.title('BMI Distribution of Diabetic Patients')
plt.xlabel('BMI')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
data_set2[data_set2.Outcome == 1].plot.scatter(x='Age', y='BMI')
```

**Variation of Blood pressure with Age**

```python
# Create a new dataframe grouped by age and blood pressure
grouped_data = data_set2.groupby('Age')['BloodPressure'].mean().reset_index()

# Create a line chart
plt.plot(grouped_data['Age'], grouped_data['BloodPressure'])
plt.xlabel('Age')
plt.ylabel('Blood Pressure')
plt.title('Blood Pressure Level by Age')
plt.show()
```

**Correlation between BMI and Glucose**

```python
correlation = data_set2['BMI'].corr(data_set2['Glucose'])
print(f"The correlation between BMI and Glucose levels is: {correlation}")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BMI', y='Glucose', data=data_set2)
plt.title('Scatter plot between BMI and Glucose levels')
plt.xlabel('BMI')
plt.ylabel('Glucose')
plt.show()

sns.heatmap(data_set2[['BMI', 'Glucose']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

**Relationship between Diabetes Pedigree Function and Age**

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='DiabetesPedigreeFunction', data=data_set2)
plt.title('Diabetes Progression vs Age')
plt.xlabel('Age')
plt.ylabel('DiabetesPedigreeFunction')
plt.show()

# Correlation
correlation = data_set2['Age'].corr(data_set2['DiabetesPedigreeFunction'])
print(f'Correlation between Age and Diabetes Pedigree Function: {correlation}')
```

**variation of Insulin levels with Age**

```python
age_bins = [0, 20, 40, 60, 80, 100]
age_labels = ['0-20', '21-40', '41-60', '61-80', '81-100']
data_set2['AgeGroup'] = pd.cut(data_set2['Age'], bins=age_bins, labels=age_labels)
```

```python
insulin_by_age = data_set2.groupby('AgeGroup')['Insulin'].mean().reset_index()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='AgeGroup', y='Insulin', data=insulin_by_age)
plt.title('Average Insulin Levels Across Different Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Average Insulin Level')
plt.show()
```
