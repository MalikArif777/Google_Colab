```python
correlation = data['serum cholestoral'].corr(data['heart disease'])
```

```python
if correlation > 0.7:
    print("The Heart Disease and Serun Cholesterol are strongly positively correlated.")
elif correlation > 0.4:
    print("The Heart Disease and Serun Cholesterol are moderately positively correlated.")
elif correlation > 0.1:
    print("The Heart Disease and Serun Cholesterol are weakly positively correlated.")
elif correlation > -0.1:
    print("The Heart Disease and Serun Cholesterol are not correlated.")
elif correlation > -0.4:
    print("The Heart Disease and Serun Cholesterol are weakly negatively correlated.")
elif correlation > -0.7:
    print("The Heart Disease and Serun Cholesterol are moderately negatively correlated.")
else:
    print("The Heart Disease and Serun Cholesterol are strongly negatively correlated.")
```

**Relation of age with heart disease**

```python
# Group by 'heart disease' and calculate the mean age
age_stats = data.groupby('heart disease')['age'].mean()
print(age_stats)
```

```python
# Plotting the age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='age', hue='heart disease', multiple='stack', kde=True)
plt.title('Age Distribution by Heart Disease Presence')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(['No Heart Disease', 'Heart Disease'])
plt.show()
```

```python
from scipy.stats import ttest_ind
# Separate the age data into two groups
age_no_disease = data[data['heart disease'] == 0]['age']
age_disease = data[data['heart disease'] == 1]['age']

# Perform an independent t-test
t_stat, p_value = ttest_ind(age_no_disease, age_disease)
print(f'T-statistic: {t_stat}, P-value: {p_value}')
```

```python
if p_value < 0.05:
    print("There is a statistically significant difference in age between patients with and without heart disease.")
else:
    print("There is no statistically significant difference in age between patients with and without heart disease.")
```

Feature Engineering

```python
# Creating a risk factor score based on cholesterol, blood pressure, and age
data['risk factor score'] = (data['serum cholestoral'] + data['resting blood pressure']) * (data['age'] / 100)
```

```python
# correlation between maximum heart rate and resting blood pressure based age

data['correlation of HR/RBP'] = data['max heart rate'] * data['resting blood pressure'] / data['age']
```
